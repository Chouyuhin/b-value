## Transform the hypoDD reloc file to the format suitable for zmap

```python
import numpy as np
import pandas as pd

# 输入和输出文件名
hypodd = pd.read_csv('/Users/chouyuhin/_Workhome/Proj_Seismicgap/QJvalidation/hypoDD_final_allevents.reloc', sep="\t")
df = hypodd[['lon','lat','year','month','day','magnitude','depth','hour','minute','second']]

# csv_file =  pd.read_csv('/Users/chouyuhin/_Workhome/Proj_Seismicgap/QJvalidation/standard_catalog_QJ_220825_230313.csv', sep=",", header=0)
# df = csv_file[['longitude','latitude','year','month','day','magnitude','depth','hour','minute','second']]
# input_file = "/Users/chouyuhin/_Workhome/Proj_Seismicgap/QJvalidation/hypoDD_final_allevents_zmap.reloc"
output_file = "/Users/chouyuhin/_Workhome/Proj_Seismicgap/allevents_zmap.txt"

# 1. 先以 object 类型读取，保留原始字符串

# data = pd.read_csv(input_file, delim_whitespace=True, header=0, dtype=float)

# # 2. 将非数字字符替换为 NaN
# data = data.applymap(lambda x: np.nan if not x.replace('.','',1).isdigit() else float(x))

# # 3. 去掉任何包含 NaN 的行
df = df.astype(float)
data = df.to_numpy()
print(data.shape)
# 从 reloc 文件中提取字段（按最常见 hypoDD 格式）
lat  = data[:, 1]
lon  = data[:, 0]
dep  = data[:, 6]

year = data[:, 2]
month = data[:, 3]
day   = data[:, 4]
hour  = data[:, 7]
minute = data[:, 8]
sec    = data[:, 9]


mag = data[:, 5]


# ZMAP 需要 decyr，如果只有年份，就先用年份代替（ZMAP 能接受
decyr = year

# 按 ZMAP 格式组合
zmap_mat = np.column_stack((lon, lat, decyr, month, day, mag, dep, hour, minute, sec))

# 保存为纯数字文件（ZMAP 可直接 load）
np.savetxt(output_file, zmap_mat, fmt="%.6f")

print("转换完成！输出文件为:", output_file)
```
